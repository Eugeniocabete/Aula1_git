#Categoriação de dados - Descoberta guiada

Crie uma coluna que classifique o cliente de acordo com o valor gasto (Income):

Até 20.000 : Baixo

Até 40.000 : Mèdio

Até 60.000 : Alto

Maior que 60.000 : Muito Alto

```python
def classificar_gasto(valor_gasto):
  if valor_gasto < 20000:
    return 'Baixo'
  elif valor_gasto < 40000:
    return 'Médio'
  elif valor_gasto < 60000:
    return 'Alto'
  elif valor_gasto >= 60000:
    return 'Muito alto'

valor = float(input('Digite o valor'))
x = classificar_gasto(valor)
x
```

```python
classificado = []
for i in df_etl['Income']:
  saida_funcao = classificar_gasto(i)
  classificado.append(saida_funcao)

print(classificado)
```

```python
df_etl['gasto_classificado'] = df_etl['Income'].apply(lambda x: classificar_gasto(x))
df_etl
```

#Categoriação de dados - Hands on

Crie uma coluna para categorizar os passageiros do titanic conforme a sua idade. Depois faça a contagem de cada categoria através do value_counts:

Até 10 anos - Criança

Até 20 anos - Jovem

Até 60 anos - Adulto

Mais de 60 anos - Idoso


```python
def classificar_idade(idade):
  if idade <= 10:
    return 'Criança'
  elif idade <= 20:
    return 'Jovem'
  elif idade <= 60:
    return 'Adulto'
  elif idade > 60:
    return 'Idoso'

df_titanic['idade_classificada'] = df_titanic['Idade'].apply(lambda x: classificar_idade(x))

df_titanic
```
